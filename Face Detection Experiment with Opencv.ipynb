```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img_raw = cv2.imread('Aranya.jpg')
```

```python
type(img_raw)
np.ndarray

img_raw.shape
```

```python
plt.imshow(img_raw)
```

```python
img = cv2.cvtColor(img_raw, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
# Loading classifier for Frontal face

haar_cascade_face = cv2.CascadeClassifier('Data/haar-cascade-files-master/haarcascade_frontalface_default.xml')
```

```python
# Face Detection using detectMultiscale module of the classifier

faces_rects = haar_cascade_face.detectMultiScale(img, scaleFactor = 1.5, minNeighbors = 5);

# Let us print the no. of faces found
print('Faces found: ', len(faces_rects))
```

```python
# loop over all the coordinates it returned and draw rectangles around them using Open CV
# drawing a red rectangle with a thickness of 40
for (x,y,w,h) in faces_rects:
     cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 40)
```
